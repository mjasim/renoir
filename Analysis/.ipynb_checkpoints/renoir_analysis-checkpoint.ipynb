```python
red = ["#D40000", "#FF0000", "#FF8C8C", "#FFC5C5"];
orange = ["#FF7D00", "#FFA854", "#FFC48C", "#FFE1C5"];
yellow = ["#FFE854", "#FFFF54", "#FFFFB1", "#FEFFDD"];
light_green = ["#00B400", "#54FF54", "#8CFF8C", "#C5FFC5"];
deep_green = ["#008000", "#009600", "#8CC68C", "#C5E2C5"];
light_blue = ["#0089E0", "#59BDFF", "#A5DBFF", "#D5EEFF"];
deep_blue = ["#0000C8", "#5151FF", "#8C8CFF", "#C5C5FF"];
purple = ["#DE00DE", "#FF54FF", "#FFC6FF", "#FFE2FF"];
achromatic = ["#000001", "#757575", "#9E9E9E", "#FFFFFF"];
```

```python
def getColor(array, array_name, value):
    if(value in array):
        return array_name + "_" + str(array.index(value) + 1)
```

```python
def getColorDist(color_dict):
    
    new_colors = {};

    for color, count in color_dict.items(): 
        color_name = getColor(red, "red", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(orange, "orange", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(yellow, "yellow", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(light_green, "light_green", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(deep_green, "deep_green", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(light_blue, "light_blue", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(deep_blue, "deep_blue", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(purple, "purple", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
        color_name = getColor(achromatic, "achromatic", color)
        if(color_name):
            if color_name not in new_colors:
                new_colors[color_name] = count
    return new_colors
```

```python
# read all int files

path = 'int/'
rev_count = 0
int_count = 0
int_len = []
rev_len = []

for filename in os.listdir(path):
    int_df = pd.read_csv(path + filename, names = ['ID','DAY','TIME','INTERACTION','COLOR'])
    #print(int_df, "\n\n")
    interactions = int_df['INTERACTION'].values.tolist()
    int_count += len(interactions)
    int_len.append(len(interactions))
    rev_count += interactions.count("revision")
    rev_len.append(interactions.count("revision"))

print("Interaction Count", int_count)
print("Revision Count", rev_count)
```

```python
#interaction stats

min_int = np.min(int_len)
max_int = np.max(int_len)
mean_int = np.mean(int_len)
std_int = np.std(int_len)

print(min_int, max_int, mean_int, std_int)

min_rev = np.min(rev_len)
max_rev = np.max(rev_len)
mean_rev = np.mean(rev_len)
std_rev = np.std(rev_len)

print(min_rev, max_rev, mean_rev, std_rev)
```

```python
def agg(color_dict):
    
    agg_dict = {"red": 0, "orange": 0, "yellow": 0, "gray": 0, "black": 0, "white": 0, "blue": 0, "green": 0, "purple": 0}
    
    for color, count in color_dict.items():
        if("red" in color):
            agg_dict["red"] += count
        elif("orange" in color):
            agg_dict["orange"] += count
        elif("yellow" in color):
            agg_dict["yellow"] += count
        elif("achromatic_2" in color):
            agg_dict["gray"] += count
        elif("achromatic_3" in color):
            agg_dict["gray"] += count    
        elif("achromatic_1" in color):
             agg_dict["black"] += count
        elif("achromatic_4" in color):
            agg_dict["white"] += count
        elif("blue" in color):
            agg_dict["blue"] += count
        elif("green" in color):
            agg_dict["green"] += count
        elif("purple" in color):
            agg_dict["purple"] += count
    return agg_dict
```
